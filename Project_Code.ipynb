## 3. Phase I Hotelling T² Control Chart (All Variables)

Iterative Phase I procedure using chi-square-based control limits to identify and remove out-of-control observations.

```python
# 1. Implement Hotelling T² function
from scipy.stats import f

def hotelling_T2_once(data, alpha=0.01):
    """
    Compute Hotelling T² statistic for each observation in Phase I data.
    
    Parameters:
    -----------
    data : pandas DataFrame of shape (m, p)
        Candidate in-control set
    alpha : float
        Significance level for control limit
    
    Returns:
    --------
    T2_values : numpy array of length m
        T² statistic for each observation
    UCL : float
        Upper control limit based on F distribution
    """
    X = data.values
    m, p = X.shape
    
    # Estimate mean and covariance
    mu_hat = X.mean(axis=0)
    S_hat = np.cov(X, rowvar=False)
    
    # Invert covariance (use pinv for numerical stability)
    try:
        S_inv = np.linalg.inv(S_hat)
    except np.linalg.LinAlgError:
        print("Warning: Covariance matrix is singular, using pseudo-inverse")
        S_inv = np.linalg.pinv(S_hat)
    
    # Compute T² for each observation
    diff = X - mu_hat
    T2 = np.einsum('ij,jk,ik->i', diff, S_inv, diff)
    
    # Phase I UCL using Montgomery formula
    F_crit = f.ppf(1 - alpha, p, m - p)
    UCL = (p * (m - 1) * (m + 1)) / (m * (m - p)) * F_crit
    
    return T2, UCL

print("Hotelling T² function defined successfully.")
```

```python
# 2. Build iterative Phase I procedure
def phase1_hotelling_T2(df, alpha=0.01, max_iterations=10):
    """
    Iterative Phase I Hotelling T² procedure.
    
    Parameters:
    -----------
    df : pandas DataFrame
        Full dataset
    alpha : float
        Significance level
    max_iterations : int
        Maximum number of iterations
    
    Returns:
    --------
    in_control_indices : list
        Final in-control observation indices
    out_of_control_info : dict
        Dictionary mapping out-of-control indices to iteration removed
    final_mu : numpy array
        Final estimated mean vector
    final_S : numpy array
        Final estimated covariance matrix
    iteration_history : list
        History of removals at each iteration
    """
    # Start with all indices
    in_control_indices = list(df.index)
    out_of_control_info = {}
    iteration_history = []
    
    print("="*70)
    print("PHASE I HOTELLING T² ITERATIVE PROCEDURE")
    print("="*70)
    print(f"Initial sample size: {len(in_control_indices)}")
    print(f"Alpha level: {alpha}")
    print(f"Maximum iterations: {max_iterations}\n")
    
    for iteration in range(1, max_iterations + 1):
        # Current in-control subset
        current_data = df.loc[in_control_indices]
        m = len(in_control_indices)
        
        print(f"Iteration {iteration}:")
        print(f"  Current in-control sample size: {m}")
        
        # Compute T² and UCL
        T2_values, UCL = hotelling_T2_once(current_data, alpha=alpha)
        
        # Identify out-of-control observations
        flagged_mask = T2_values > UCL
        n_flagged = np.sum(flagged_mask)
        
        print(f"  UCL = {UCL:.4f}")
        print(f"  Max T² = {T2_values.max():.4f}")
        print(f"  Flagged observations: {n_flagged}")
        
        if n_flagged == 0:
            print(f"  → No new observations flagged. Convergence reached.")
            iteration_history.append({
                'iteration': iteration,
                'n_in_control': m,
                'n_flagged': 0,
                'UCL': UCL
            })
            break
        
        # Remove flagged observations
        flagged_indices = [in_control_indices[i] for i, flag in enumerate(flagged_mask) if flag]
        for idx in flagged_indices:
            out_of_control_info[idx] = iteration
        
        in_control_indices = [idx for idx in in_control_indices if idx not in flagged_indices]
        
        iteration_history.append({
            'iteration': iteration,
            'n_in_control': m,
            'n_flagged': n_flagged,
            'UCL': UCL,
            'flagged_indices': flagged_indices
        })
        
        print(f"  → Removed {n_flagged} observations")
        print(f"  → Remaining in-control: {len(in_control_indices)}\n")
    
    # Compute final parameters
    final_data = df.loc[in_control_indices]
    final_mu = final_data.values.mean(axis=0)
    final_S = np.cov(final_data.values, rowvar=False)
    
    print("="*70)
    print("FINAL RESULTS:")
    print(f"  In-control observations: {len(in_control_indices)}")
    print(f"  Out-of-control observations: {len(out_of_control_info)}")
    print(f"  Total iterations: {len(iteration_history)}")
    print("="*70)
    
    return in_control_indices, out_of_control_info, final_mu, final_S, iteration_history

# Run the Phase I procedure
in_control_idx, out_of_control_info, final_mu, final_S, history = phase1_hotelling_T2(df, alpha=0.01, max_iterations=10)
```

```python
# 3. Summarize results by iteration
print("\n" + "="*70)
print("SUMMARY BY ITERATION")
print("="*70)

iteration_summary = []
for item in history:
    iteration_summary.append({
        'Iteration': item['iteration'],
        'In-Control': item['n_in_control'],
        'Flagged': item['n_flagged'],
        'UCL': f"{item['UCL']:.4f}"
    })

summary_df = pd.DataFrame(iteration_summary)
print(summary_df.to_string(index=False))

# Analyze which iteration removed the most
removal_counts = {}
for idx, iter_num in out_of_control_info.items():
    removal_counts[iter_num] = removal_counts.get(iter_num, 0) + 1

print(f"\n" + "="*70)
print("OUT-OF-CONTROL REMOVALS BY ITERATION")
print("="*70)

if len(removal_counts) > 0:
    for iter_num in sorted(removal_counts.keys()):
        print(f"  Iteration {iter_num}: {removal_counts[iter_num]} observations removed")
    
    most_removals_iter = max(removal_counts, key=removal_counts.get)
    print(f"\nMost observations removed in Iteration {most_removals_iter}: {removal_counts[most_removals_iter]} observations")
else:
    print("  No observations were removed - all observations are in-control!")
```

```python
# 4. Plot the final T² chart
print("\n" + "="*70)
print("COMPUTING FINAL T² VALUES FOR ALL OBSERVATIONS")
print("="*70)

# Compute T² for all 552 observations using final parameters
X_all = df.values
diff_all = X_all - final_mu

# Use pseudo-inverse for stability
try:
    S_inv = np.linalg.inv(final_S)
except np.linalg.LinAlgError:
    print("Using pseudo-inverse for final covariance matrix")
    S_inv = np.linalg.pinv(final_S)

T2_all = np.einsum('ij,jk,ik->i', diff_all, S_inv, diff_all)

# Compute final UCL
m_final = len(in_control_idx)
p = df.shape[1]
F_crit = f.ppf(1 - 0.01, p, m_final - p)
UCL_final = (p * (m_final - 1) * (m_final + 1)) / (m_final * (m_final - p)) * F_crit

print(f"Final UCL: {UCL_final:.4f}")
print(f"Max T² (all obs): {T2_all.max():.4f}")
print(f"Max T² (in-control): {T2_all[in_control_idx].max():.4f}")

if len(out_of_control_info) > 0:
    print(f"Max T² (out-of-control): {T2_all[list(out_of_control_info.keys())].max():.4f}")
else:
    print(f"Max T² (out-of-control): N/A (no out-of-control observations)")

# Create the plot
fig, ax = plt.subplots(figsize=(16, 7))

# Plot in-control points
ax.scatter(in_control_idx, T2_all[in_control_idx], 
           c='blue', alpha=0.6, s=30, label='In-Control', zorder=2)

# Plot out-of-control points (if any)
out_of_control_indices = list(out_of_control_info.keys())
if len(out_of_control_indices) > 0:
    ax.scatter(out_of_control_indices, T2_all[out_of_control_indices], 
               c='red', alpha=0.8, s=50, marker='X', label='Out-of-Control', zorder=3)

# Plot UCL
ax.axhline(y=UCL_final, color='red', linestyle='--', linewidth=2, 
           label=f'UCL = {UCL_final:.2f}', zorder=1)

# Formatting
ax.set_xlabel('Observation Index', fontsize=12, fontweight='bold')
ax.set_ylabel('Hotelling T² Statistic', fontsize=12, fontweight='bold')
ax.set_title('Phase I Hotelling T² Control Chart (All Observations)', 
             fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='upper right', fontsize=10)
ax.grid(True, alpha=0.3)
ax.set_xlim(-10, 560)

# Add text box with summary
textstr = f'In-Control: {len(in_control_idx)}\nOut-of-Control: {len(out_of_control_indices)}\nAlpha: 0.01'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
ax.text(0.02, 0.98, textstr, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()
```

```python
# Compute T² for all observations using chi-square-based approach
from scipy.stats import chi2

print("="*70)
print("CHI-SQUARE-BASED HOTELLING T² ANALYSIS")
print("="*70)

# Use all 552 observations as candidate in-control set
X = df.values
m, p = X.shape

print(f"\nDataset dimensions: {m} observations × {p} variables")

# Compute sample mean and covariance
mu_hat = X.mean(axis=0)
S_hat = np.cov(X, rowvar=False)

# Use pseudoinverse for numerical stability
S_inv = np.linalg.pinv(S_hat)

print(f"Sample mean vector shape: {mu_hat.shape}")
print(f"Sample covariance matrix shape: {S_hat.shape}")
print(f"Using pseudoinverse for covariance matrix inversion")

# Compute T² for all observations
diff = X - mu_hat
T2_all = np.einsum('ij,jk,ik->i', diff, S_inv, diff)

print(f"\nT² statistics computed for all {m} observations")
print(f"T² range: [{T2_all.min():.4f}, {T2_all.max():.4f}]")
print(f"T² mean: {T2_all.mean():.4f}")
print(f"T² median: {np.median(T2_all):.4f}")
```

### Iterative Refinement with 5 Iterations

Implementing iterative procedure that removes OOC observations and refits the model for cleaner reference dataset.

```python
import os

# Create output folder for plots
output_folder = 'Phase_I_Plots'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"Created folder: {output_folder}")
else:
    print(f"Folder already exists: {output_folder}")
    
print(f"All plots will be saved to: {os.path.abspath(output_folder)}")
```

```python
# Iterative Phase I Refinement with Chi-Square T²
def iterative_phase1_refinement(df, alpha=0.01, max_iterations=5):
    """
    Iterative Phase I refinement using chi-square-based Hotelling T².
    
    Parameters:
    -----------
    df : pandas DataFrame
        Full dataset
    alpha : float
        Significance level
    max_iterations : int
        Maximum number of iterations
    
    Returns:
    --------
    results : dict
        Complete results for each iteration and final parameters
    """
    # Initialize
    current_indices = list(df.index)
    all_ooc_history = {}  # Maps index to iteration when removed
    iteration_results = []
    
    print("="*80)
    print("ITERATIVE PHASE I REFINEMENT - CHI-SQUARE T²")
    print("="*80)
    print(f"Initial dataset size: {len(current_indices)} observations")
    print(f"Number of variables: {df.shape[1]}")
    print(f"Significance level (α): {alpha}")
    print(f"Maximum iterations: {max_iterations}\n")
    
    for iteration in range(1, max_iterations + 1):
        print(f"\n{'='*80}")
        print(f"ITERATION {iteration}")
        print(f"{'='*80}")
        
        # Get current in-control data
        df_current = df.loc[current_indices]
        n_current = len(current_indices)
        p = df.shape[1]
        
        print(f"Current sample size: {n_current} observations")
        
        # Standardize current data
        scaler_current = StandardScaler()
        X_current = df_current.values
        X_standardized = scaler_current.fit_transform(X_current)
        
        # Compute mean and covariance
        mu_current = X_current.mean(axis=0)
        S_current = np.cov(X_current, rowvar=False)
        
        # Invert covariance (use pseudoinverse for stability)
        try:
            S_inv_current = np.linalg.inv(S_current)
        except np.linalg.LinAlgError:
            print("  → Using pseudoinverse (matrix singular)")
            S_inv_current = np.linalg.pinv(S_current)
        
        # Compute T² for current observations
        diff_current = X_current - mu_current
        T2_current = np.einsum('ij,jk,ik->i', diff_current, S_inv_current, diff_current)
        
        # Chi-square UCL
        UCL_chi = chi2.ppf(1 - alpha, df=p)
        
        # Identify out-of-control points
        ooc_mask = T2_current > UCL_chi
        n_ooc = np.sum(ooc_mask)
        
        # Get actual indices of OOC points
        ooc_indices_iter = [current_indices[i] for i, flag in enumerate(ooc_mask) if flag]
        
        print(f"Chi-square UCL (df={p}, α={alpha}): {UCL_chi:.4f}")
        print(f"T² range: [{T2_current.min():.4f}, {T2_current.max():.4f}]")
        print(f"Out-of-control observations flagged: {n_ooc} ({n_ooc/n_current*100:.2f}%)")
        
        if n_ooc > 0:
            print(f"OOC T² range: [{T2_current[ooc_mask].min():.4f}, {T2_current[ooc_mask].max():.4f}]")
            print(f"OOC indices: {sorted(ooc_indices_iter)}")
        
        # Store iteration results
        iteration_results.append({
            'iteration': iteration,
            'n_observations': n_current,
            'n_ooc': n_ooc,
            'ooc_indices': ooc_indices_iter,
            'UCL': UCL_chi,
            'T2_values': T2_current.copy(),
            'observation_indices': current_indices.copy(),
            'mu': mu_current.copy(),
            'S': S_current.copy()
        })
        
        # Record which iteration removed each OOC point
        for idx in ooc_indices_iter:
            all_ooc_history[idx] = iteration
        
        # Check for convergence
        if n_ooc == 0:
            print(f"\n{'='*80}")
            print(f"✓ CONVERGENCE REACHED: No outliers detected in iteration {iteration}")
            print(f"{'='*80}")
            break
        
        # Remove OOC points for next iteration
        current_indices = [idx for idx in current_indices if idx not in ooc_indices_iter]
        print(f"→ Removed {n_ooc} observations")
        print(f"→ Remaining for next iteration: {len(current_indices)} observations")
    
    # Final in-control set
    final_ic_indices = current_indices
    final_mu = iteration_results[-1]['mu']
    final_S = iteration_results[-1]['S']
    
    print(f"\n{'='*80}")
    print("FINAL PHASE I RESULTS")
    print(f"{'='*80}")
    print(f"Total iterations performed: {len(iteration_results)}")
    print(f"Final in-control observations: {len(final_ic_indices)}")
    print(f"Total out-of-control removed: {len(all_ooc_history)}")
    print(f"Percentage removed: {len(all_ooc_history)/len(df)*100:.2f}%")
    
    return {
        'iteration_results': iteration_results,
        'final_ic_indices': final_ic_indices,
        'all_ooc_history': all_ooc_history,
        'final_mu': final_mu,
        'final_S': final_S,
        'alpha': alpha,
        'final_ic_count': len(final_ic_indices)
    }

# Run iterative refinement for both alpha levels
print("="*100)
print("RUNNING ITERATIVE REFINEMENT FOR BOTH SIGNIFICANCE LEVELS")
print("="*100)
print("\nComparing conservative (α=0.01) vs. sensitive (α=0.05) detection\n")

print("\n╔" + "="*98 + "╗")
print("║" + " "*35 + "ANALYSIS 1: α = 0.01 (99%)" + " "*37 + "║")
print("╚" + "="*98 + "╝\n")

refinement_results_001 = iterative_phase1_refinement(df, alpha=0.01, max_iterations=5)

print("\n\n╔" + "="*98 + "╗")
print("║" + " "*35 + "ANALYSIS 2: α = 0.05 (95%)" + " "*37 + "║")
print("╚" + "="*98 + "╝\n")

refinement_results_005 = iterative_phase1_refinement(df, alpha=0.05, max_iterations=5)

# Store default reference (α=0.01)
refinement_results = refinement_results_001
```

## 4. PCA-Based Phase I Procedure (Two-Stage Approach)

Implementing PCA-based monitoring with both α=0.01 and α=0.05 significance levels.

**Step 4: Initial Screening (All Data)**
- Standardize and apply PCA on all 552 observations
- Retain k=64 PCs (90% variance threshold)
- Compute T² statistic in PC space with F-based UCL
- Compute Q-statistic (SPE) for residuals using Jackson-Mudholkar UCL
- Flag observations where T² > UCL_T² OR Q > UCL_Q

**Step 5: IC Refinement & Final Model**
- Remove flagged observations from Step 4
- Refit StandardScaler and PCA on IC subset only
- Determine k_final based on 90% variance (IC data)
- Transform all observations using IC-based model
- Compute final T² and Q with IC-based control limits
- Final classification of all observations

```python
# PCA-based Phase I Procedure (Step 4 & Step 5)
from scipy.stats import f, chi2, norm
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd

def pca_phase1_step4(df, k, alpha=0.01, verbose=True):
    """
    Step 4: Initial T² and Q on all data with specified k PCs.
    
    Parameters:
    -----------
    df : DataFrame
        Dataset (n × p)
    k : int
        Number of principal components to retain
    alpha : float
        Significance level
    verbose : bool
        Print detailed output
        
    Returns:
    --------
    results : dict
        Step 4 results including T², Q, flags, and PCA model
    """
    if verbose:
        print("="*80)
        print("STEP 4: HOTELLING'S T² AND Q (SPE) – PCA-BASED")
        print("="*80)
        print(f"Dataset: {df.shape[0]} observations × {df.shape[1]} variables")
        print(f"Number of PCs to retain: k = {k}")
        print(f"Significance level α: {alpha}")
        print()
    
    X = df.values
    n, p = X.shape
    
    # Standardize (z-score)
    scaler = StandardScaler()
    Z = scaler.fit_transform(X)
    
    # PCA on standardized data
    pca = PCA()
    pca.fit(Z)
    
    # Eigenvalues (variance of each PC)
    lambdas = pca.explained_variance_
    
    # Full PCA scores
    T_full = pca.transform(Z)
    
    # Split into retained and residual PCs
    T_k = T_full[:, :k]
    lambda_k = lambdas[:k]
    
    if k < p:
        T_resid = T_full[:, k:]
        lambda_resid = lambdas[k:]
    else:
        T_resid = np.zeros((n, 0))
        lambda_resid = np.array([])
    
    # Compute Hotelling T² in PC space: T²_i = sum_j (t_ij² / lambda_j)
    T2 = np.sum((T_k**2) / lambda_k, axis=1)
    
    # Phase I T² control limit (F-approximation)
    # UCL_T² = k*(n-1)/(n-k) * F_{k,n-k}(1-α)
    UCL_T2_F = k * (n - 1) / (n - k) * f.ppf(1 - alpha, k, n - k)
    
    # Chi-square approximate limit for reference
    UCL_T2_chi = chi2.ppf(1 - alpha, df=k)
    
    if verbose:
        print(f"T² UCL (F-based, alpha={alpha}): {UCL_T2_F:.3f}")
        print(f"T² UCL (Chi-square approx, df={k}): {UCL_T2_chi:.3f}")
    
    # Flag T² outliers (using F-based limit)
    flag_T2 = T2 > UCL_T2_F
    
    # Compute Q-statistic (Squared Prediction Error): Q_i = sum over residual PCs of t_ij²
    if T_resid.shape[1] > 0:
        Q = np.sum(T_resid**2, axis=1)
        
        # Jackson–Mudholkar Q limit
        theta1 = np.sum(lambda_resid)
        theta2 = np.sum(lambda_resid**2)
        theta3 = np.sum(lambda_resid**3)
        
        # Avoid division by zero if residual part is tiny
        if theta2 > 0 and theta3 > 0:
            h0 = 1 - (2 * theta1 * theta3) / (3 * theta2**2)
            z_alpha = norm.ppf(1 - alpha)
            
            Q_term = (
                (z_alpha * np.sqrt(2 * theta2 * h0**2)) / theta1
                + 1
                + (theta2 * h0 * (h0 - 1)) / (theta1**2)
            )
            UCL_Q = theta1 * (Q_term**(1 / h0))
        else:
            UCL_Q = np.max(Q) * 1.1
        
        if verbose:
            print(f"Q-statistic UCL (alpha={alpha}): {UCL_Q:.3f}")
        flag_Q = Q > UCL_Q
    else:
        Q = np.zeros(n)
        UCL_Q = np.nan
        flag_Q = np.array([False]*n)
        if verbose:
            print("No residual PCs beyond k; Q-statistic not meaningful (set to 0).")
    
    # Combine flags
    flag_any = flag_T2 | flag_Q
    
    if verbose:
        print(f"\nNumber of observations flagged by T²: {flag_T2.sum()}")
        print(f"Number of observations flagged by Q: {flag_Q.sum()}")
        print(f"Number of observations flagged by either: {flag_any.sum()}")
        print("\n" + "="*80)
    
    return {
        'n': n,
        'p': p,
        'k': k,
        'scaler': scaler,
        'pca': pca,
        'T2': T2,
        'Q': Q,
        'UCL_T2': UCL_T2_F,
        'UCL_T2_chi': UCL_T2_chi,
        'UCL_Q': UCL_Q,
        'flag_T2': flag_T2,
        'flag_Q': flag_Q,
        'flag_any': flag_any,
        'lambdas': lambdas,
        'alpha': alpha
    }

def pca_phase1_step5(df, step4_results, variance_threshold=0.90, alpha=0.01, verbose=True):
    """
    Step 5: Robust IC selection & Final PCA model.
    Refit scaler and PCA on IC subset, then evaluate all observations.
    
    Parameters:
    -----------
    df : DataFrame
        Original dataset (n × p)
    step4_results : dict
        Results from Step 4
    variance_threshold : float
        Target cumulative variance for choosing k_final
    alpha : float
        Significance level
    verbose : bool
        Print detailed output
        
    Returns:
    --------
    results : dict
        Final results including IC-based classifications
    """
    if verbose:
        print("="*80)
        print("STEP 5: ROBUST IC SELECTION & FINAL PCA MODEL")
        print("="*80)
        print(f"Target cumulative variance: {variance_threshold*100:.0f}%")
        print(f"Significance level α: {alpha}")
        print()
    
    X = df.values
    n, p = X.shape
    
    # Define in-control (IC) set based on Step 4
    flag_any_prev = step4_results['flag_any']
    ic_mask = ~flag_any_prev
    ic_indices = np.where(ic_mask)[0]
    ooc_indices = np.where(~ic_mask)[0]
    
    n_ic = ic_mask.sum()
    
    if verbose:
        print(f"Number of outliers in Step 4 (flag_any): {flag_any_prev.sum()}")
        print(f"Number of IC observations (used for refit): {n_ic}")
        print(f"Number of OOC observations removed: {(~ic_mask).sum()}")
        print()
    
    # Standardize based on IC data
    X_ic = X[ic_mask]
    scaler_ic = StandardScaler()
    Z_ic = scaler_ic.fit_transform(X_ic)
    
    # Apply same scaling to ALL data (including OOC) using IC mean/std
    Z_all = scaler_ic.transform(X)
    
    # PCA on IC standardized data
    pca_ic = PCA()
    pca_ic.fit(Z_ic)
    
    lambdas_ic = pca_ic.explained_variance_
    ev_ic = pca_ic.explained_variance_ratio_
    cum_ev_ic = np.cumsum(ev_ic)
    
    # Choose k as smallest number of PCs with cumulative variance >= variance_threshold
    k_final = int(np.argmax(cum_ev_ic >= variance_threshold) + 1)
    
    if verbose:
        print(f"Chosen k_final = {k_final} PCs (IC model) for >= {variance_threshold*100:.0f}% variance")
        print(f"Cumulative variance explained: {cum_ev_ic[k_final-1]*100:.2f}%")
        print()
    
    # Compute T² and Q for ALL observations under IC-based PCA model
    T_all = pca_ic.transform(Z_all)
    
    # Split scores into retained and residual parts
    T_k_all = T_all[:, :k_final]
    lambda_k_ic = lambdas_ic[:k_final]
    
    if k_final < p:
        T_resid_all = T_all[:, k_final:]
        lambda_resid_ic = lambdas_ic[k_final:]
    else:
        T_resid_all = np.zeros((n, 0))
        lambda_resid_ic = np.array([])
    
    # T² for all observations (using IC model)
    T2_final = np.sum((T_k_all**2) / lambda_k_ic, axis=1)
    
    # F-based UCL for T² (Phase I style, but using IC count)
    UCL_T2_F_final = k_final * (n_ic - 1) / (n_ic - k_final) * f.ppf(1 - alpha, k_final, n_ic - k_final)
    UCL_T2_chi_final = chi2.ppf(1 - alpha, df=k_final)
    
    if verbose:
        print(f"T² UCL (IC, F-based, alpha={alpha}): {UCL_T2_F_final:.3f}")
        print(f"T² UCL (IC, Chi-square approx, df={k_final}): {UCL_T2_chi_final:.3f}")
    
    flag_T2_final = T2_final > UCL_T2_F_final
    
    # Q-statistic for all observations
    if T_resid_all.shape[1] > 0:
        Q_final = np.sum(T_resid_all**2, axis=1)
        
        theta1 = np.sum(lambda_resid_ic)
        theta2 = np.sum(lambda_resid_ic**2)
        theta3 = np.sum(lambda_resid_ic**3)
        
        if theta2 > 0 and theta3 > 0:
            h0 = 1 - (2 * theta1 * theta3) / (3 * theta2**2)
            z_alpha = norm.ppf(1 - alpha)
            
            Q_term = (
                (z_alpha * np.sqrt(2 * theta2 * h0**2)) / theta1
                + 1
                + (theta2 * h0 * (h0 - 1)) / (theta1**2)
            )
            UCL_Q_final = theta1 * (Q_term**(1 / h0))
        else:
            UCL_Q_final = np.max(Q_final) * 1.1
        
        if verbose:
            print(f"Q UCL (IC, alpha={alpha}): {UCL_Q_final:.3f}")
        flag_Q_final = Q_final > UCL_Q_final
    else:
        Q_final = np.zeros(n)
        UCL_Q_final = np.nan
        flag_Q_final = np.array([False]*n)
        if verbose:
            print("No residual PCs; Q-statistic not meaningful.")
    
    flag_any_final = flag_T2_final | flag_Q_final
    
    n_ooc_T2_final = flag_T2_final.sum()
    n_ooc_Q_final = flag_Q_final.sum()
    n_ooc_any_final = flag_any_final.sum()
    n_ooc_both_final = (flag_T2_final & flag_Q_final).sum()
    
    if verbose:
        print(f"\nFinal IC-based model results:")
        print(f"  Flagged by T² only: {n_ooc_T2_final - n_ooc_both_final} ({(n_ooc_T2_final - n_ooc_both_final)/n*100:.2f}%)")
        print(f"  Flagged by Q only: {n_ooc_Q_final - n_ooc_both_final} ({(n_ooc_Q_final - n_ooc_both_final)/n*100:.2f}%)")
        print(f"  Flagged by both: {n_ooc_both_final} ({n_ooc_both_final/n*100:.2f}%)")
        print(f"  Total flagged (T² OR Q): {n_ooc_any_final} ({n_ooc_any_final/n*100:.2f}%)")
        print(f"  Total IC: {n - n_ooc_any_final} ({(n - n_ooc_any_final)/n*100:.2f}%)")
        print("\n" + "="*80)
    
    return {
        'n': n,
        'p': p,
        'k': k_final,
        'n_ic': n_ic,
        'ic_mask': ic_mask,
        'ic_indices': ic_indices,
        'ooc_indices': ooc_indices,
        'scaler': scaler_ic,
        'pca': pca_ic,
        'T2': T2_final,
        'Q': Q_final,
        'UCL_T2': UCL_T2_F_final,
        'UCL_T2_chi': UCL_T2_chi_final,
        'UCL_Q': UCL_Q_final,
        'flag_T2': flag_T2_final,
        'flag_Q': flag_Q_final,
        'flag_any': flag_any_final,
        'n_ooc_T2': n_ooc_T2_final,
        'n_ooc_Q': n_ooc_Q_final,
        'n_ooc_any': n_ooc_any_final,
        'n_ooc_both': n_ooc_both_final,
        'ooc_T2_indices': np.where(flag_T2_final)[0],
        'ooc_Q_indices': np.where(flag_Q_final)[0],
        'ooc_any_indices': np.where(flag_any_final)[0],
        'ic_final_indices': np.where(~flag_any_final)[0],
        'lambdas': lambdas_ic,
        'cum_var': cum_ev_ic[k_final-1],
        'alpha': alpha,
        'variance_threshold': variance_threshold,
        'step4_flag_any': flag_any_prev
    }

print("PCA-based Phase I functions (Step 4 & Step 5) defined successfully.")
```

## 6. Contribution Plot Analysis

Contribution plots help identify which specific variables are responsible for observations being flagged as out-of-control. This root cause analysis is essential for understanding the underlying reasons for anomalies and taking corrective actions.

### What are Contribution Plots?

- **T² Contributions**: Show how much each original variable contributes to the Hotelling T² statistic
- **Q Contributions**: Show how much each variable contributes to the Q (SPE) statistic in PCA
- **Chi-Square T² Contributions**: Show variable contributions to the chi-square-based T² statistic

### Approach:
1. Identify top outliers from each method
2. Compute variable-level contributions for each outlier
3. Visualize contributions to identify root causes

### 6.1 Chi-Square T² Method - Contribution Analysis

```python
# Chi-Square T² Contribution Analysis (α=0.01)
# For the chi-square based T², each variable's contribution is calculated as:
# Contribution_i = (x_i - μ_i)² / σ_i² for each variable

print("="*100)
print("CHI-SQUARE T² CONTRIBUTION ANALYSIS (α=0.01)")
print("="*100)

# Get the final in-control indices from the iterative refinement (α=0.01)
ic_indices_chi = refinement_results_001['final_ic_indices']
ooc_indices_chi = refinement_results_001['all_ooc_history']

print(f"\nTotal out-of-control observations: {len(ooc_indices_chi)}")
print(f"Analyzing top 5 outliers with highest T² values\n")

# Compute T² values for all observations using final IC parameters
X_all = df[[col for col in df.columns]].values
mu_final = refinement_results_001['final_mu']
S_final = refinement_results_001['final_S']
S_inv_final = np.linalg.inv(S_final)

# Compute T² for all observations
T2_all_chi = np.array([
    (X_all[i] - mu_final) @ S_inv_final @ (X_all[i] - mu_final)
    for i in range(len(X_all))
])

# Get top 5 outliers by T² value
top_indices_chi = np.argsort(T2_all_chi)[-5:][::-1]

print("Top 5 observations with highest Chi-Square T² values:")
for i, idx in enumerate(top_indices_chi, 1):
    print(f"  {i}. Observation {idx}: T² = {T2_all_chi[idx]:.2f}")

# Function to compute chi-square T² contributions
def compute_chi_T2_contributions(obs_vector, mu, S_inv):
    """
    Compute contribution of each variable to chi-square T² statistic.
    
    For multivariate T², the contribution of variable i is:
    contrib_i = (obs - mu)' * S_inv[:, i] * (obs - mu)[i]
    
    This shows how much each variable contributes to the overall T² statistic.
    """
    diff = obs_vector - mu
    p = len(obs_vector)
    contributions = np.zeros(p)
    
    for i in range(p):
        # Contribution from variable i
        contributions[i] = diff[i] * np.dot(S_inv[i, :], diff)
    
    return contributions

# Create contribution plots for top 5 outliers
print("\nGenerating contribution plots...")

for idx in top_indices_chi:
    obs_vector = X_all[idx]
    contributions = compute_chi_T2_contributions(obs_vector, mu_final, S_inv_final)
    
    # Get top 10 contributing variables
    top_vars = np.argsort(np.abs(contributions))[-10:][::-1]
    
    # Create plot
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 10))
    
    # Plot 1: All variables
    colors = ['red' if c > 0 else 'blue' for c in contributions]
    ax1.bar(range(len(contributions)), contributions, color=colors, alpha=0.6, edgecolor='black')
    ax1.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax1.set_xlabel('Variable Index', fontsize=12)
    ax1.set_ylabel('Contribution to T²', fontsize=12)
    ax1.set_title(f'Chi-Square T² Contributions for Observation {idx} (T² = {T2_all_chi[idx]:.2f})', 
                  fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Top 10 contributing variables (zoomed)
    ax2.bar(range(10), contributions[top_vars], 
            color=['red' if c > 0 else 'blue' for c in contributions[top_vars]], 
            alpha=0.7, edgecolor='black')
    ax2.set_xticks(range(10))
    ax2.set_xticklabels([f'{df.columns[v]}' for v in top_vars], rotation=45, ha='right')
    ax2.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax2.set_xlabel('Variable Name', fontsize=12)
    ax2.set_ylabel('Contribution to T²', fontsize=12)
    ax2.set_title(f'Top 10 Contributing Variables (Observation {idx})', fontsize=13, fontweight='bold')
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    save_path = os.path.join(output_folder, f'Chi_T2_contribution_obs_{idx}.png')
    plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.close()
    
    print(f"  ✓ Saved: {save_path}")
    
    # Print top contributing variables
    print(f"\n  Top 5 contributing variables for Observation {idx}:")
    for i, var_idx in enumerate(top_vars[:5], 1):
        print(f"    {i}. {df.columns[var_idx]}: {contributions[var_idx]:.4f}")

print(f"\n{'='*100}")
print(f"Chi-Square T² contribution plots saved to: {output_folder}/")
print(f"{'='*100}")
```

### 6.2 PCA-Based Method - T² and Q Contribution Analysis

```python
# PCA T² and Q Contribution Analysis (α=0.01)
print("="*100)
print("PCA-BASED CONTRIBUTION ANALYSIS (α=0.01)")
print("="*100)

# Get Step 5 results (IC-based model)
ic_mask_pca = step5_results_001['ic_mask']
T2_final_pca = step5_results_001['T2']
Q_final_pca = step5_results_001['Q']
flag_T2_pca = step5_results_001['flag_T2']
flag_Q_pca = step5_results_001['flag_Q']

# Get top outliers for T² and Q
top_T2_indices = np.argsort(T2_final_pca)[-5:][::-1]
top_Q_indices = np.argsort(Q_final_pca)[-5:][::-1]

print(f"\nTop 5 observations with highest PCA T² values:")
for i, idx in enumerate(top_T2_indices, 1):
    status = "OOC" if flag_T2_pca[idx] else "IC"
    print(f"  {i}. Observation {idx}: T² = {T2_final_pca[idx]:.2f} ({status})")

print(f"\nTop 5 observations with highest PCA Q values:")
for i, idx in enumerate(top_Q_indices, 1):
    status = "OOC" if flag_Q_pca[idx] else "IC"
    print(f"  {i}. Observation {idx}: Q = {Q_final_pca[idx]:.2f} ({status})")

# Re-fit PCA model on IC data for contribution analysis
X_ic_pca = df.iloc[ic_mask_pca].values
scaler_pca = StandardScaler()
Z_ic_pca = scaler_pca.fit_transform(X_ic_pca)
Z_all_pca = scaler_pca.transform(df.values)

# Fit PCA
pca_model = PCA()
pca_model.fit(Z_ic_pca)

# Get PCA components
k_pca = step5_results_001['k']
loadings_pca = pca_model.components_.T  # shape (p, p)
lambdas_pca = pca_model.explained_variance_
T_scores_all = pca_model.transform(Z_all_pca)  # shape (n, p)

print(f"\nPCA Model: k = {k_pca} principal components retained")
print(f"Variance explained: {step5_results_001['cum_var']:.2%}")

# Function to compute PCA T² contributions
def compute_pca_T2_contributions(t_scores, loadings, lambdas, k):
    """
    Compute contribution of each original variable to PCA T² statistic.
    
    T² contribution from variable j = sum_i=1^k [(t_i * loading_ji)² / lambda_i]
    where t_i is the score on PC i, and loading_ji is the loading of variable j on PC i.
    """
    p = loadings.shape[0]
    contributions = np.zeros(p)
    
    for j in range(p):
        for i in range(k):
            contributions[j] += (t_scores[i] * loadings[j, i])**2 / lambdas[i]
    
    return contributions

# Function to compute PCA Q contributions
def compute_pca_Q_contributions(obs_z, T_scores, loadings, k):
    """
    Compute contribution of each variable to Q (SPE) statistic.
    
    Q contribution from variable j = (obs_z_j - reconstructed_j)²
    where reconstructed_j = sum_i=1^k [t_i * loading_ji]
    """
    p = len(obs_z)
    reconstructed = np.dot(T_scores[:k], loadings[:, :k].T)
    residual = obs_z - reconstructed
    contributions = residual**2
    
    return contributions

# Generate T² contribution plots
print(f"\n{'='*100}")
print("GENERATING PCA T² CONTRIBUTION PLOTS")
print(f"{'='*100}\n")

for idx in top_T2_indices:
    t_scores = T_scores_all[idx, :]
    obs_z = Z_all_pca[idx, :]
    
    contributions = compute_pca_T2_contributions(t_scores, loadings_pca, lambdas_pca, k_pca)
    
    # Get top 10 contributing variables
    top_vars = np.argsort(contributions)[-10:][::-1]
    
    # Create plot
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 10))
    
    # Plot 1: All variables
    ax1.bar(range(len(contributions)), contributions, color='steelblue', alpha=0.6, edgecolor='black')
    ax1.set_xlabel('Variable Index', fontsize=12)
    ax1.set_ylabel('Contribution to T²', fontsize=12)
    ax1.set_title(f'PCA T² Contributions for Observation {idx} (T² = {T2_final_pca[idx]:.2f})', 
                  fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Top 10 contributing variables
    ax2.bar(range(10), contributions[top_vars], color='steelblue', alpha=0.7, edgecolor='black')
    ax2.set_xticks(range(10))
    ax2.set_xticklabels([f'{df.columns[v]}' for v in top_vars], rotation=45, ha='right')
    ax2.set_xlabel('Variable Name', fontsize=12)
    ax2.set_ylabel('Contribution to T²', fontsize=12)
    ax2.set_title(f'Top 10 Contributing Variables (Observation {idx})', fontsize=13, fontweight='bold')
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    save_path = os.path.join(output_folder, f'PCA_T2_contribution_obs_{idx}.png')
    plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.close()
    
    print(f"  ✓ Saved: {save_path}")
    
    # Print top contributing variables
    print(f"  Top 5 contributing variables for Observation {idx}:")
    for i, var_idx in enumerate(top_vars[:5], 1):
        print(f"    {i}. {df.columns[var_idx]}: {contributions[var_idx]:.4f}")
    print()

# Generate Q contribution plots
print(f"{'='*100}")
print("GENERATING PCA Q CONTRIBUTION PLOTS")
print(f"{'='*100}\n")

for idx in top_Q_indices:
    t_scores = T_scores_all[idx, :]
    obs_z = Z_all_pca[idx, :]
    
    contributions = compute_pca_Q_contributions(obs_z, t_scores, loadings_pca, k_pca)
    
    # Get top 10 contributing variables
    top_vars = np.argsort(contributions)[-10:][::-1]
    
    # Create plot
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 10))
    
    # Plot 1: All variables
    ax1.bar(range(len(contributions)), contributions, color='coral', alpha=0.6, edgecolor='black')
    ax1.set_xlabel('Variable Index', fontsize=12)
    ax1.set_ylabel('Contribution to Q', fontsize=12)
    ax1.set_title(f'PCA Q Contributions for Observation {idx} (Q = {Q_final_pca[idx]:.2f})', 
                  fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Top 10 contributing variables
    ax2.bar(range(10), contributions[top_vars], color='coral', alpha=0.7, edgecolor='black')
    ax2.set_xticks(range(10))
    ax2.set_xticklabels([f'{df.columns[v]}' for v in top_vars], rotation=45, ha='right')
    ax2.set_xlabel('Variable Name', fontsize=12)
    ax2.set_ylabel('Contribution to Q', fontsize=12)
    ax2.set_title(f'Top 10 Contributing Variables (Observation {idx})', fontsize=13, fontweight='bold')
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    save_path = os.path.join(output_folder, f'PCA_Q_contribution_obs_{idx}.png')
    plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.close()
    
    print(f"  ✓ Saved: {save_path}")
    
    # Print top contributing variables
    print(f"  Top 5 contributing variables for Observation {idx}:")
    for i, var_idx in enumerate(top_vars[:5], 1):
        print(f"    {i}. {df.columns[var_idx]}: {contributions[var_idx]:.4f}")
    print()

print(f"{'='*100}")
print(f"All PCA contribution plots saved to: {output_folder}/")
print(f"{'='*100}")
```

### 6.3 Contribution Analysis Summary

**Key Insights from Contribution Plots:**

1. **Chi-Square T² Contributions**:
   - Shows which variables contribute most to the overall Hotelling T² statistic
   - Positive contributions indicate variables deviating in the positive direction
   - Negative contributions indicate variables deviating in the negative direction
   - Useful for identifying root causes of multivariate anomalies

2. **PCA T² Contributions**:
   - Shows which original variables contribute most through the retained principal components
   - Captures contributions in the main variance space (k retained PCs)
   - Higher contributions indicate variables with strong deviations in the principal component space

3. **PCA Q Contributions**:
   - Shows which variables contribute to the residual (unexplained) variation
   - Captures unusual patterns not explained by the main PCs
   - Useful for detecting novel types of anomalies or process changes

**Interpretation Guidelines:**
- Variables with high contributions are the primary drivers of out-of-control signals
- Multiple variables with moderate contributions suggest multivariate correlation effects
- Focus corrective actions on variables with consistently high contributions across multiple outliers
